```python
# Import required libraries

import pandas as pd
import numpy as np
```

```python
# Import train data
train_plus=pd.read_csv("./NSL-KDD/Text_files/train/KDDTrain+.txt",header=None)
train_20=pd.read_csv("./NSL-KDD/Text_files/train/KDDTrain+_20Percent.txt",header=None)
```

```python
# Import Test data
test_plus=pd.read_csv("./NSL-KDD/Text_files/test/KDDTest+.txt",header=None)
test_21=pd.read_csv("./NSL-KDD/Text_files/test/KDDTest-21.txt",header=None)
```

```python
test_plus.head()
```

```python
# preprocess data
train=pd.concat([train_plus,train_20])
```

```python
train.head()
```

```python
train=train.rename({41: "Attack"}, axis='columns')
train.loc[train["Attack"] == 'normal', "Attack"] = 0
train.loc[train["Attack"] != 0, "Attack"] = 1
```

```python
train.head()
```

```python

X=train.drop(['Attack'],axis=1)
y=train['Attack']
```

```python
X
```

```python
# Leave_one_out_encoding
import category_encoders
ce=category_encoders.leave_one_out.LeaveOneOutEncoder(cols=[1,2,3])
```

```python
X=ce.fit_transform(X,y)
```

```python
X
```

```python
# InterQuartile Scaling
from sklearn.preprocessing import RobustScaler
transformer = RobustScaler().fit(X)
```

```python
X_scaled=transformer.transform(X)
```

```python
X=pd.DataFrame(X_scaled,columns=X.columns)
```

```python
X.head()
```

```python
y.loc[y == 0] = '0'
y.loc[y == 1] = '1'
```

```python
# Model Trainig 
from sklearn import tree
DecisionTree_model = tree.DecisionTreeClassifier()
DecisionTree_model.fit(X, y)
```

```python
# Preprocess test data

test_plus=test_plus.rename({41: "Attack"}, axis='columns')
test_plus.loc[test_plus["Attack"] == 'normal', "Attack"] = 0
test_plus.loc[test_plus["Attack"] != 0, "Attack"] = 1
test_plus.head()
```

```python
X_test=test_plus.drop(['Attack'],axis=1)
y_test=test_plus['Attack']
```

```python
# Leave_one_out_encoding
import category_encoders
ce=category_encoders.leave_one_out.LeaveOneOutEncoder(cols=[1,2,3])
X_test=ce.fit_transform(X_test,y_test)
# InterQuartile Scaling
from sklearn.preprocessing import RobustScaler
transformer = RobustScaler().fit(X_test)
X_test_scaled=transformer.transform(X_test)
X_test=pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

```python
y_test.loc[y_test == 0] = '0'
y_test.loc[y_test == 1] = '1'
```

```python
# Evaluation
y_pred_DecisionTree=DecisionTree_model.predict(X_test)
```

```python
from sklearn import metrics 
Conf_matrix_DecisionTree=metrics.confusion_matrix(y_test, y_pred_DecisionTree)
TP_DecisionTree=Conf_matrix_DecisionTree[0,0]
FN_DecisionTree=Conf_matrix_DecisionTree[1,0]
FP_DecisionTree=Conf_matrix_DecisionTree[0,1]
TN_DecisionTree=Conf_matrix_DecisionTree[1,1]

TPR_DecisionTree=TP_DecisionTree / float(TP_DecisionTree + FN_DecisionTree)
FNR_DecisionTree=FN_DecisionTree / float(TP_DecisionTree + FN_DecisionTree)
PPV_DecisionTree=TP_DecisionTree / float(TP_DecisionTree + FP_DecisionTree)
```

```python
print('Decision Tree True positive rate (TPR, recall) on the test+ dataset: ', TPR_DecisionTree)
print("\n")
print('Decision Tree False negative rate (FNR) on the test+ dataset: ', FNR_DecisionTree)
print("\n")
print('Decision Tree positive predictive value (PPR, precision) on the test+ dataset: ', PPV_DecisionTree)
print("\n")
print('Decision Tree Accuracy on the test+ dataset: ', metrics.accuracy_score(y_pred_DecisionTree,y_test))
print("\n")
```
